#### Groupby - Any groupby operation involves one of the following operations on the original object. They are −
1 Splitting the Object 
2 Applying a function 
3 Combining the results 

```python
grp=avacd.groupby("size")["nb_sold"].sum()                    # groupby(argument)[index]
print(grp)
```

#### Note : Matplotlib is the whole package; matplotlib. pyplot is a module in matplotlib; and pylab is a module that gets installed alongside matplotlib . ... pylab is a convenience module that bulk imports matplotlib. pyplot (for plotting) and numpy (for mathematics and working with arrays) in a single name space.

#### Bar Chart - Bar charts are great for comparing two or more values that are related categorically, for example, a bar chart comparing of small, large and extra large avocados sold.

```python
import matplotlib.pyplot as plt                         
grp.plot(kind="bar")                           # Previously the code was (from matplotlib import pyplot)
plt.show()                                             
```

```python
avacd.head()
```

```python
groupdt=avacd.groupby("date")["nb_sold"].sum()
groupdt
```

#### Line Graph - Line charts are ideal for showing data trends and patterns over time ..... When smaller changes exist, line graphs are better to use than bar graphs

```python
groupdt.plot(kind="line")
plt.show()                                       
```

### Scatter plots - are used to observe relationship between variables and uses dots to represent the relationship between them.  Scatter plots are widely used to represent relation among variables and how change in one affects the other

```python
scatterplt=avacd.plot(x="nb_sold", y="avg_price",  kind="scatter", title="number of avocados sold vs average price")
```

```python
avacd.head()
```

```python
avacd.hist()
plt.show()
```

```python
avacd[avacd["type"]=="conventional"]["avg_price"].hist()
avacd[avacd["type"]=="organic"]["avg_price"].hist()
plt.legend(["conventional","organic"])
plt.show
```

```python
avacd[avacd["type"]=="conventional"]["avg_price"].hist(alpha=0.7)
avacd[avacd["type"]=="organic"]["avg_price"].hist(alpha=0.7)
plt.legend(["conventional","organic"])
plt.show
```

```python
avacd[avacd["type"]=="conventional"]["avg_price"].hist(alpha=0.75,bins=20)
avacd[avacd["type"]=="organic"]["avg_price"].hist(alpha=0.75, bins=20)
plt.legend(["conventional","organic"])
plt.show
```
